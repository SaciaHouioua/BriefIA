## Régression linéaire

Un modèle de regression linéaire est un modèle qui cherche à établir ou à trouver une relation entre une variable, dite expliquée "y", et une ou plusieurs variables, dites explicatives "x". 
Parmi les modèles de régression linéaire, le plus simple est celui de l'ajustument affine qui consiste à déterminer la droite  " y = ax + b" approchant au mieux un nuage de points dans le plan.

```python

from sklearn.linear_model import LinearRegression
```

```python
# redimensionner les tableaux
y=data["Selling_Price"].values.reshape(y.shape[0],1)
X = data["Year"].values.reshape(y.shape[0],1)
print(y.shape)
print(X.shape)
```

```python
model = LinearRegression()
model.fit(X, y)
S = model.score(X,y)
print("Score =", S)
model.predict
model.coef_
slope = model.coef_[0]
intercept = model.intercept_
print(slope)
print(intercept)
```

```python

plt.scatter(data["Year"], y, color='blue', label='numpy Linear Regression') # Nuage de points
plt.plot(X, slope*X + intercept, color="red", label='fitted line' )
plt.legend()
```

## Support Vector Machines (SVM)

Les machines à vecteurs de support ou séparateur à vaste marge (en anglais support vector machine, SVM) est une solution à un problème de classification le but est de trouver une frontière optimale entre deux ou plusieurs catégories. Cette technique est largement utilisée pour classer des email ou des documents, identifier des cancers,..  
De prime abort le problème au quel nous sommes confronté ne relève pas d'une classification simple.
Par contre, nous pouvons imaginer une solution en combinant dans les deux principes :
classification et regression.
commencer par déterminer des catégories, puis faire une regression linéaire multivariable, polynomiale de degré 2 ou 3.
