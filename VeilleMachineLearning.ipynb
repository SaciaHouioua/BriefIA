# Avec Sklearn (la fonction stats.linregress(x,y))

Pour faire du Machine Learning avec SKLEARN, la démarche est toujours la même quelque soit le modèle que nous voulons implenter. L'ensemble des modèles et des alogrithmes ont été implémentée selon une architecture orientée objet et chaque modèle dispose de sa propre classe. Pour créer un modèle on génère un objet de la classe correspondante.
Une fois le modèle initialisé, on est en mesure de l'entrainer, l'évaluer, et l'utiliser à l'aide de trois méthodes cummunent à toutes les classes SKLEARN.
L'interface d'utlisation est la suivante   
model = ....  
model.fit(X,y)  
model.score(X,y)  
model.predict(X,y)  


```python

from sklearn.linear_model import LinearRegression
```

```python
y=data["Selling_Price"].values.reshape(y.shape[0],1)
X = data["Year"].values.reshape(y.shape[0],1)
print(y.shape)
print(X.shape)
```

```python
model = LinearRegression()
model.fit(X, y)
S = model.score(X,y)
print("Score =", S)
model.predict
model.coef_
slope = model.coef_[0]
intercept = model.intercept_
print(slope)
print(intercept)
```

```python

plt.scatter(data["Year"], y, color='blue', label='numpy Linear Regression') # Nuage de points
plt.plot(X, slope*X + intercept, color="red", label='fitted line' )
plt.legend()
```

Nous  remarquons que les trois graphes sont identiques car nous obtenons les mêmes valeurs avec les trois méthodes.
